```python
import getpass
from pathlib import Path

from autoscholar.parser.llama_parser import LlamaParser, ParsingMode, ResultType
```

```python
llama_parse_api = getpass.getpass("Enter your LlamaParse API key: ")
```

```python
import nest_asyncio

nest_asyncio.apply()

test_pdf_path = Path("test-pdf.pdf")
parser = LlamaParser(
    api_key=llama_parse_api,
    parsing_mode=ParsingMode.BALANCED,
    result_type=ResultType.MD,
)

result = parser.parse(test_pdf_path)
```

```python
from io import BytesIO
from pprint import pprint

from PIL import Image

pprint(result.get_markdown_documents()[0].text[:100])

pprint(result.get_image_names())

Image.open(BytesIO(result.get_image_data("img_p0_1.png")))
```
